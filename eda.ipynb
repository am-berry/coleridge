```python
def basic_checker(num):
    # Load the json in
    # Check labs against cleaned json
    count = 0
    train = pd.read_csv('train.csv')
    labs = set(train['cleaned_label'].tolist())
    train_dir = './train/'
    files = os.listdir(train_dir)[:num]
    for fi in files:
        filename = os.path.splitext(fi)[0]
        true_lab = train[train['Id']==filename]['cleaned_label'].tolist()
        with open(train_dir+fi) as f:
            d = json.load(f)
        long_text = ""
        for dic in d:
            for k, v in dic.items():
                long_text += k
                long_text += '$$$$$'
                long_text += v
                long_text += '$$$$$'
        lt = clean_text(long_text)
        count += any(l in lt for l in labs)

    return count, count/num
```

```python
basic_checker(100)
```
